```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame de ejemplo con valores NaN
datos = np.random.rand(100, 800)
datos[np.random.randint(0, 100, 100), np.random.randint(0, 800, 100)] = np.nan
df = pd.DataFrame(datos)

# Crear una mscara booleana donde True representa valores NaN
mascara_nan = df.isna()

# Configurar colores para los valores NaN y no NaN
colores = ['#FFFFFF', '#1f77b4']  # Blanco para NaN, azul para otros valores

# Graficar la matriz de colores
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(),  linewidths=0.5, linecolor='gray', cmap='viridis')
plt.title('Matriz de Colores (NaN en Blanco)')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.show()
```
