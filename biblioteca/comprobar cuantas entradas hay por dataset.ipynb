```python
import pandas as pd
import os

# Ruta de la carpeta que contiene los archivos CSV
carpeta = r"C:\Users\arzua\OneDrive\Escritorio\caracterizacion\PTU"

# Lista para almacenar los DataFrames de cada archivo CSV
dataframes = []
os.listdir(carpeta)

for archivo in os.listdir(carpeta):
    if archivo.endswith('.CSV'):  # Asegúrate de que sean archivos CSV
        # Construye la ruta completa al archivo
        ruta_completa = os.path.join(carpeta, archivo)
        
        # Lee el archivo CSV y crea un DataFrame
        df = pd.read_csv(ruta_completa,encoding='latin1')
        
        # Almacena el DataFrame en la lista
        dataframes.append(df)
    else:
        print('no llega aqui')
# Concatena todos los DataFrames en uno solo
dataframe_completo = pd.concat(dataframes, ignore_index=True)
#dataframes
dataframe_completo
#dataframe_completo['Date']
# Convertir la columna "fecha" al formato de fecha
dataframe_completo['Date'] = pd.to_datetime(dataframe_completo['Date'][1:],format='%d.%m.%Y')
dataframe_completo.info()
#dataframes
```

```python
#Crear un data frame de todos los naT para cuantificar su dimension.
df_nan_date = dataframe_completo[pd.isna(dataframe_completo['Date'])]
```

```python
#quiero contar cuantos hay por fechas iguales y cuantos dias hay.


# Supongamos que tienes una serie temporal 'fechas' en formato de fecha
# fechas = pd.to_datetime(df['fecha'])

# Si tienes una serie temporal de ejemplo
fechas = pd.date_range('2024-01-01', '2024-01-05', freq='D').append(pd.date_range('2024-01-01', '2024-01-03', freq='D'))

# Cuenta el número de ocurrencias de cada fecha
conteo_por_fecha = fechas.value_counts().sort_index()

# Calcula el número total de días únicos
num_dias_unicos = len(conteo_por_fecha)

# Imprime el resultado
print("Número de días únicos:", num_dias_unicos)
print("\nNúmero de registros por día:")
print(conteo_por_fecha)
```
